# Statistics for Consumers with Fraud

```python
fig2, axs2 = plt.subplots(2, 2)
fig2.suptitle('Statistics for Consumers with Fraud', fontsize=18)
plt.subplots_adjust(hspace=0.8)

data.loc[0].plot(ax=axs2[0, 0], color='firebrick', grid=True)
axs2[0, 0].set_title('Consumption of Consumer 0', fontsize=16)
axs2[0, 0].set_xlabel('Dates of Consumption')
axs2[0, 0].set_ylabel('Consumption')

data.loc[0].hist(color='firebrick', ax=axs2[0, 1], grid=True)
axs2[0, 1].set_title('Histogram', fontsize=16)
axs2[0, 1].set_xlabel('Values')
axs2[0, 1].set_ylabel('Frequency')

data.loc[0].plot.kde(color='firebrick', ax=axs2[1, 0], grid=True)
axs2[1, 0].set_title('Density Estimation', fontsize=16)
axs2[1, 0].set_xlabel('Values')
axs2[1, 0].set_ylabel('Density')

data.loc[0].describe().drop(['count']).plot(kind='bar', ax=axs2[1, 1], color='firebrick', grid=True)
axs2[1, 1].set_title('Statistics', fontsize=16)
axs2[1, 1].set_ylabel('Values')
```

# 	L'Estimation de Densit montre la distribution continue des donnes et semble indiquer une distribution bimodale, avec deux types de valeurs de consommation distincts.=>Par exemple, un pic plus bas pourrait correspondre   une consommation d'lectricit relativement faible, tandis qu'un pic plus lev pourrait reprsenter une consommation d'lectricit beaucoup plus leve.

# 	Les Statistiques sous forme de graphique en barres indiquent que, si on exclut les valeurs extrmes (comme le max), les autres mesures descriptives (moyenne, cart-type, mdiane) sont relativement basses.=>la majorit des valeurs de consommation se situent  des niveaux plus faibles

# Consommation des quatres semaines 

```python
fig4, axs4 = plt.subplots(2, 1)
fig4.suptitle('Four Week Consumption', fontsize=16)
plt.subplots_adjust(hspace=0.5)
for i in range(59, 83, 7):
    axs4[0].plot(data.iloc[0, i:i + 7].to_numpy(), marker='>', linestyle='-', label='$week {i}$'.format(i=(i % 58) % 6))
axs4[0].legend(loc='best')
axs4[0].set_title('Without Fraud', fontsize=14)
axs4[0].set_ylabel('Consumption')
axs4[0].grid(True)

for i in range(59, 83, 7):
    axs4[1].plot(data.iloc[4, i:i + 7].to_numpy(), marker='>', linestyle='-', label='$week {i}$'.format(i=(i % 58) % 6))
axs4[1].legend(loc='best')
axs4[1].set_title('With Fraud', fontsize=14)
axs4[1].set_ylabel('Consumption')
axs4[1].grid(True)
```

# Les graphiques comparent la consommation sur quatre semaines pour un consommateur sans fraude et un avec fraude.

# Cohrence des tendances hebdomadaires : Pour le consommateur sans fraude, les tendances hebdomadaires semblent plus cohrentes et stables d'une semaine  l'autre. Cela suggre que ce consommateur maintient des habitudes de consommation relativement constantes et prvisibles au fil du temps.

# Variabilit accrue pour le consommateur avec fraude : En revanche, pour le consommateur avec fraude, il y a une plus grande variabilit dans les tendances hebdomadaires. On observe des fluctuations plus importantes dans la consommation d'une semaine  l'autre. Cette variabilit accrue peut tre le rsultat de comportements irrguliers associs  des activits frauduleuses telles que le dtournement d'lectricit.

# Impact sur la dtection de fraude : Cette diffrence de variabilit entre les deux consommateurs peut tre utilise comme indicateur potentiel de fraude. Une consommation qui varie considrablement d'une semaine  l'autre, sans motif vident, peut tre suspecte et mriter une enqute approfondie pour dterminer s'il y a effectivement une activit frauduleuse en jeu.


```python
fig5, axs5 = plt.subplots(1, 2)
a = []
for i in range(59, 83, 7):
    a.append(data.iloc[0, i:i + 7].to_numpy())
cor = pd.DataFrame(a).transpose().corr()
cax = axs5[0].matshow(cor)
for (i, j), z in np.ndenumerate(cor):
    axs5[0].text(j, i, '{:0.1f}'.format(z), ha='center', va='center', color='white')
alpha = ['week 1', 'week 2', 'week 3', 'week 4']
axs5[0].set_xticklabels([''] + alpha)
axs5[0].set_yticklabels([''] + alpha)
axs5[0].set_title('Customer without Fraud', fontsize=16)

a = []
for i in range(59, 83, 7):
    a.append(data.iloc[4, i:i + 7].to_numpy())
cor = pd.DataFrame(a).transpose().corr()
cax = axs5[1].matshow(cor)
for (i, j), z in np.ndenumerate(cor):
    axs5[1].text(j, i, '{:0.1f}'.format(z), ha='center', va='center', color='white')
axs5[1].set_xticklabels([''] + alpha)
axs5[1].set_yticklabels([''] + alpha)
axs5[1].set_title('Customer with Fraud', fontsize=16)
fig5.colorbar(cax)
```

#  montre qu'il existe une forte corrlation entre les donnes de consommation d'lectricit des clients normaux, c'est--dire que la plupart des valeurs de PCC sont suprieures  0,7 (une valeur de PCC proche de 1 signifie une corrlation plus forte) . ==> ce qui peut indiquer un modle de consommation cohrent.
