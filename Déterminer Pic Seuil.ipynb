# Méthode de Tukey (Interquartile Range, IQR)

```python
# Charger les données
data = pd.read_csv(r"S:\Mes documents\Bureau\PFE MID\Les articles\ElectricityTheftDetection\SmartGridFraudDetection-master\SmartGridFraudDetection-master\data\preprocessedR90.csv")
data = data.dropna()  # Assurez-vous que les données sont propres


# Supposons que les colonnes de consommation commencent après 'CONS_NO'
consumption_columns = data.columns[2:]

# Calcul des quartiles et de l'IQR
Q1 = data[consumption_columns].quantile(0.25).values.flatten().mean()
Q3 = data[consumption_columns].quantile(0.75).values.flatten().mean()
IQR = Q3 - Q1

# Définir le seuil
seuil_pic = Q3 + 1.5 * IQR

print(f"Le seuil calculé pour les pics est de: {seuil_pic}")
```

# Validation Croisée pour Tester Plusieurs Seuils

```python
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier

# Charger les données
data = pd.read_csv(r"S:\Mes documents\Bureau\PFE MID\Les articles\ElectricityTheftDetection\SmartGridFraudDetection-master\SmartGridFraudDetection-master\data\preprocessedR90.csv")
data = data.dropna()  # Assurez-vous que les données sont propres


# Supposons que les colonnes de consommation commencent après 'CONS_NO'
consumption_columns = data.columns[2:]

# Séparer les caractéristiques et les étiquettes
X = data[consumption_columns]
y = data['FLAG']  # Supposons que 'FLAG' est la colonne d'étiquette

# Définir une gamme de seuils
seuils = np.linspace(0.5, 0.8, 10)

best_score = 0
best_seuil = seuils[0]

# Validation croisée pour tester chaque seuil
for seuil in seuils:
    X_temp = (X > seuil).astype(int)  # Binariser les données en fonction du seuil
    clf = DecisionTreeClassifier()
    scores = cross_val_score(clf, X_temp, y, cv=5)
    mean_score = scores.mean()
    if mean_score > best_score:
        best_score = mean_score
        best_seuil = seuil

print(f"Le meilleur seuil est {best_seuil} avec un score moyen de {best_score}")
```

# Modèle de Mélange Gaussien

```python
import pandas as pd
import numpy as np
from sklearn.mixture import GaussianMixture

# Charger les données
df = pd.read_csv(r"S:\Mes documents\Bureau\PFE MID\Les articles\ElectricityTheftDetection\SmartGridFraudDetection-master\SmartGridFraudDetection-master\data\preprocessedR90.csv")
df = df.dropna()  # Assurez-vous que les données sont propres

# Supposons que les colonnes de consommation commencent après 'CONS_NO'
consumption_columns = df.columns[2:]

# Ajuster un modèle de mélange gaussien
gmm = GaussianMixture(n_components=2)  # On suppose deux composantes
gmm.fit(df[consumption_columns].values.flatten().reshape(-1, 1))

# Déterminer le seuil basé sur les composantes
threshold = np.max(gmm.means_)

print(f"Le seuil calculé pour les pics est de: {threshold}")
```
